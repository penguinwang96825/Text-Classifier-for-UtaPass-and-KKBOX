```python
import feedparser
from datetime import datetime, timedelta
```

```python
comment_info = {
    "author": "", 
    "title": "",
    "score": 0.0,
    "date": "",
    "comment": "",
    "version": 0.0
    }

page = 1
total_pages = 10
UtaPass_urls = []
    
for url in range(total_pages):
    urls = "https://itunes.apple.com/jp/rss/customerreviews/page=" + str(page) + "/id=579510737/sortby=mostrecent/xml"
    print(urls)
    UtaPass_urls.append(urls)
    page += 1
```

```python
UtaPass_urls[9]
```

```python
n = 0
while n < 10:
    feed = feedparser.parse(UtaPass_urls[n])

    items = []
    review_count = 0
    for entry in feed.entries[1:]:
        # get comment's author name
        comment_info["author"] = entry['author']
        # get comment's title
        comment_info["title"] = entry['title']
        # get comment's score
        comment_info["score"] = entry['im_rating']
        # get comment date
        comment_info["date"] = datetime.strptime(entry['updated'], "%Y-%m-%dT%H:%M:%S-07:00")
        # get comments
        comment_info['comment'] = entry['content'][0]['value']
        # get version
        comment_info['version'] = entry['im_version']

        comment_info = {
            "title": comment_info["title"],
            "score": comment_info["score"],
            "date": comment_info["date"],
            "comment": comment_info['comment'],
            "version": comment_info['version']
        }
        items.append(comment_info)

        review_count += 1
    
    import pandas as pd
    df = pd.DataFrame(items, columns=["author", "title", "score", "date", "comment", "version"])
    
    n += 1
```
